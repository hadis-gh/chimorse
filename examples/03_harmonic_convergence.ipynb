## Harmonic Convergence
Inspect how many harmonics truncation for fourier expansion capture the data dependencies over the angles. ($h_{\chi}, h_{\psi}$)

---

### Data Exploration & Model Parameter Choice

```python
import matplotlib.pyplot as plt

from chimorse.config import (
    FigureContext, get_colors, PLOT_PARAMS, MOLECULES
)
from chimorse.dataio import load_data
from chimorse.plotting import (
    plot_convergence_panel, 
    plot_harmonic_convergence
)

plt.rcParams.update(PLOT_PARAMS)
```

```python
molecule_name = 'PA'
interaction = 'EP'
zero_zeta = True

R = 9

molecule = MOLECULES[molecule_name]

df_org = load_data(molecule, interaction, zero_zeta)
colors = get_colors()

ctx_org = FigureContext(base="Figures", molecule=molecule.name, data_type="org", interaction=interaction)
```

### Harmonic Convergence

```python
data_type = 'D'
harmonics_chi_ranges = {'EP':(1, 9), 'EA': (1,9), 'OP':(13,21), 'OA':(13,21)}
harmonics_psi = 3

plot_convergence_panel(df_org, interaction, screw_step=molecule.screw_step, 
                       left_label= True, colors=colors, data_type=data_type,
                       harmonics=(harmonics_chi_ranges[interaction], harmonics_psi), 
                       error_text_size=7, save_path=ctx_org.dir())
```

```python
plot_harmonic_convergence(df_org, interaction, screw_step = molecule.screw_step, 
                          harmonics=(harmonics_chi_ranges[interaction], harmonics_psi))
```
