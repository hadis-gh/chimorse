## Radial Analysis — Curve Fit $E(r)$


Fit 1D radial potential models (Morse, Lennard-Jones) to $E(r)$ at a fixed orientation and compare against the raw data.

---

### Data Exploration & Model Parameter Choice

```python
import matplotlib.pyplot as plt

from chimorse.config import (
    FigureContext, get_colors, PLOT_PARAMS, MOLECULES
)
from chimorse.dataio import load_data
from chimorse.models import Morse_1D, LennardJones_1D
from chimorse.fitting import fit_Er_1D_lmfit

from chimorse.plotting import plot_ER_model_compare

plt.rcParams.update(PLOT_PARAMS)
```

```python
molecule_name = 'PA'
interaction = 'EP'
zero_zeta = True

molecule = MOLECULES[molecule_name]

df_org = load_data(molecule, interaction, zero_zeta)
colors = get_colors()

ctx_org = FigureContext(base="Figures", molecule=molecule.name, data_type="org", interaction=interaction)
```

### Fit Morse vs. LJ to line profiles $E(r)$

```python
phi_vals = (300, 12)
r_range_small = (7.5, 11)
e_range_small = (-1.5, 0)

potential_models = [
    {'name': 'Morse', 'df': fit_Er_1D_lmfit(df_org, phi_vals, Morse_1D, [.2, 10., 1.], fit_mode='fit')},
    {'name': 'Lennard-Jones','df': fit_Er_1D_lmfit(df_org, phi_vals, LennardJones_1D, [.2, 9.], fit_mode='fit')}
    ]

plot_ER_model_compare(df_org, potential_models, phi_vals=phi_vals, 
                    # r_range=r_range_small, e_range=e_range_small,
                    save_path=ctx_org.dir())
```
